```python
# æ—¶é—´featureæž„é€ 
uk['Date'] = pd.to_datetime(uk['Date'],format='%d/%m/%Y')
uk['Time'] = pd.to_datetime(uk['Time'],format='%H:%M')

uk['Year'] = uk['Date'].dt.year

nmonth = uk['Date'].dt.month
uk['Month'] = nmonth

uk['Day'] = uk['Date'].dt.day

days_each_month = np.cumsum(np.array([0,31,28,31,30,31,30,31,31,30,31,30,31]))
nday = [days_each_month[i-1] for i in nmonth.values]
nday = nday + uk['Date'].dt.day.values
uk['Day_of_year'] = nday

uk['Hour'] = uk['Time'].dt.hour

uk['Minute_of_day']=uk['Hour']*60.0+uk['Time'].dt.minute

uk['Time']=uk['Time'].dt.time
```

```python
uk.reset_index(drop=True, inplace=True)
# uk.to_csv('UK_2015-2019_with-nan.csv')
```

```python
uk.info()
```

```python
Nan_list=['Casualty_IMD_Decile','Driver_IMD_Decile','Vehicle_IMD_Decile','Propulsion_Code','Road_Surface_Conditions',
          'Junction_Location','1st_Point_of_Impact','Casualty_Home_Area_Type']
for i in Nan_list:
    uk[i]=uk[i].replace(np.nan,13)
    
median_list= ['Age_of_Driver','Age_of_Casualty','Engine_Capacity_(CC)','Age_of_Vehicle']
for i in median_list:
    uk[i+'_Na'] = 0
    uk.loc[uk[i].isnull(),i+'_Na'] = 1
    uk[i] = uk[i].fillna(uk[i].median())
uk['Age_Band_of_Driver'] = uk['Age_Band_of_Driver'].fillna(7)
uk['Age_Band_of_Casualty'] = uk['Age_Band_of_Casualty'].fillna(6)   

from scipy import stats
mode_list= ['Car_Passenger','Special_Conditions_at_Site','Carriageway_Hazards','Towing_and_Articulation','Skidding_and_Overturning',
            '1st_Point_of_Impact','Was_Vehicle_Left_Hand_Drive?','Hit_Object_in_Carriageway','Vehicle_Location-Restricted_Lane',
          'Vehicle_Leaving_Carriageway','Hit_Object_off_Carriageway','Driver_Home_Area_Type']
for i in mode_list:
    uk[i+'_Na'] = 0
    uk.loc[uk[i].isnull(),i+'_Na'] = 1
    uk[i] = uk[i].fillna(stats.mode(uk[i])[0][0])  
    
trans_list=uk.columns.tolist()
remove_list=['Accident_Index','Date','Time','Longitude','Latitude']
for i in remove_list:
    trans_list.remove(i)
uk[trans_list] = uk[trans_list].astype('int')
```
